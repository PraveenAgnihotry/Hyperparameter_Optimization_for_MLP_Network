```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the datasets
train_df = pd.read_excel('THA2train.xlsx')
validation_df = pd.read_excel('THA2validate.xlsx')
```

```python
# Define activation functions and loss functions
activationFunctions = {
    'sigmoid': (lambda x: 1 / (1 + np.exp(-x)),
                lambda x: (1 / (1 + np.exp(-x))) * (1 - (1 / (1 + np.exp(-x))))),
    'relu': (lambda x: np.maximum(0, x), lambda x: np.where(x > 0, 1, 0)),
    'tanh': (lambda x: np.tanh(x), lambda x: 1.0 - np.tanh(x)**2),
    'softmax': (lambda x: np.exp(x) / np.sum(np.exp(x), axis=1, keepdims=True), None)
}
```

```python
lossFunctions = {
    'cross_entropy': lambda y, y_hat: -np.sum(y * np.log(y_hat + 1e-9)) / y.shape[0],
    'cross_entropy_derivative': lambda y, y_hat: y_hat - y
}
```

```python
# Parse data
def parse_data(df):
    X = np.array(df[['X_0', 'X_1']].values)  # Feature columns
    y = np.zeros((X.shape[0], 2))  # One-hot encode the labels (2 classes)
    for i in range(X.shape[0]):
        y[i, int(df['y'].values[i])] = 1
    X = X / np.max(X, axis=0)  # Normalize features
    return X, y
```

```python
# MLP model definition
class MLP:
    def __init__(self, input_neurons, hidden_layer_neurons, output_neurons, hidden_activation='relu',
                 output_activation='softmax', loss_function='cross_entropy', learning_rate=0.001,
                 batch_size=64, epochs=5000, patience=100, dropout_rate=0.0001):
        self.input_neurons = input_neurons
        self.hidden_layer_neurons = hidden_layer_neurons  # List of neurons per hidden layer
        self.output_neurons = output_neurons
        self.hidden_activation, self.hidden_activation_derivative = activationFunctions[hidden_activation]
        self.output_activation, _ = activationFunctions[output_activation]
        self.loss_function = lossFunctions[loss_function]
        self.loss_function_derivative = lossFunctions[f"{loss_function}_derivative"]
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.epochs = epochs
        self.patience = patience
        self.dropout_rate = dropout_rate

        # Initialize weights and biases
        self.weights, self.biases = self._initialize_weights()

        # Training progress storage
        self.losses = []
        self.val_losses = []

    # Initializaing weights and bias
    def _initialize_weights(self):
        layers = [self.input_neurons] + self.hidden_layer_neurons + [self.output_neurons]
        weights = []
        biases = []
        for i in range(len(layers) - 1):
            weight = np.random.randn(layers[i], layers[i + 1]) * np.sqrt(2 / layers[i])
            bias = np.zeros((1, layers[i + 1]))
            weights.append(weight)
            biases.append(bias)
        return weights, biases

    # Forward pass
    def forward(self, X):
        activations = [X]
        z_values = []
        for i in range(len(self.weights) - 1):
            z = np.dot(activations[-1], self.weights[i]) + self.biases[i]
            a = self.hidden_activation(z)
            # Apply dropout during training
            if self.dropout_rate > 0:
                dropout_mask = (np.random.rand(*a.shape) > self.dropout_rate).astype(np.float32)
                a *= dropout_mask / (1 - self.dropout_rate)
            z_values.append(z)
            activations.append(a)

        z = np.dot(activations[-1], self.weights[-1]) + self.biases[-1]
        a = self.output_activation(z)
        z_values.append(z)
        activations.append(a)
        return activations, z_values

    # Backward pass
    def backward(self, X, y, activations, z_values):
        gradients_w = [0] * len(self.weights)
        gradients_b = [0] * len(self.biases)

        delta = self.loss_function_derivative(y, activations[-1])
        for i in reversed(range(len(self.weights))):
            gradients_w[i] = np.dot(activations[i].T, delta)
            gradients_b[i] = np.sum(delta, axis=0, keepdims=True)
            if i > 0:
                delta = np.dot(delta, self.weights[i].T) * self.hidden_activation_derivative(z_values[i - 1])
        return gradients_w, gradients_b

    # Updating weights and biases
    def update_parameters(self, gradients_w, gradients_b):
        for i in range(len(self.weights)):
            self.weights[i] -= self.learning_rate * gradients_w[i]
            self.biases[i] -= self.learning_rate * gradients_b[i]

    # Training the model
    def train(self, X_train, y_train, muted=True, X_val=None, y_val=None):
        best_val_loss = float('inf')
        epochs_no_improve = 0
        for epoch in range(self.epochs):
            indices = np.random.permutation(len(X_train))
            X_train, y_train = X_train[indices], y_train[indices]

            for i in range(0, len(X_train), self.batch_size):
                X_batch = X_train[i:i + self.batch_size]
                y_batch = y_train[i:i + self.batch_size]
                activations, z_values = self.forward(X_batch)
                gradients_w, gradients_b = self.backward(X_batch, y_batch, activations, z_values)
                self.update_parameters(gradients_w, gradients_b)

            activations_train, _ = self.forward(X_train)
            loss = self.loss_function(y_train, activations_train[-1])
            self.losses.append(loss)

            if X_val is not None and y_val is not None:
                activations_val, _ = self.forward(X_val)
                val_loss = self.loss_function(y_val, activations_val[-1])
                self.val_losses.append(val_loss)
                if val_loss < best_val_loss:
                    best_val_loss = val_loss
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                if epochs_no_improve >= self.patience:
                    if not muted:
                        print(f"Early stopping at epoch {epoch + 1}")
                    break

            if not muted and epoch % 100 == 0:
                val_loss_msg = f", Validation Loss: {val_loss:.4f}" if X_val is not None else ""
                print(f"Epoch {epoch + 1}/{self.epochs}, Training Loss: {loss:.4f}{val_loss_msg}")

    # To make predictions
    def predict(self, X):
        activations, _ = self.forward(X)
        return np.argmax(activations[-1], axis=1)

    # Plotting the loss curve
    def plot_loss(self):
        plt.figure(figsize=(10, 6))
        plt.plot(range(len(self.losses)), self.losses, label="Training Loss", linewidth=2)
        if self.val_losses:
            plt.plot(range(len(self.val_losses)), self.val_losses, label="Validation Loss", linewidth=2, linestyle='--')
        plt.xlabel("Epochs", fontsize=12)
        plt.ylabel("Loss", fontsize=12)
        plt.title("Training and Validation Loss Curve", fontsize=14, fontweight='bold')
        plt.legend(fontsize=10)
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()

    # To find evaluation accuracy
    def evaluate_accuracy(self, X, y, dataset_name="Dataset"):
        predictions = self.predict(X)
        true_labels = np.argmax(y, axis=1)
        accuracy = np.mean(predictions == true_labels)
        print(f"{dataset_name} Accuracy: {accuracy:.2f}")
        return accuracy
```

```python
# Parse the datasets
X_train, y_train = parse_data(train_df)
X_val, y_val = parse_data(validation_df)
```

```python
# Create a model with zero-initialized weights
learning_rate = 0.0001
epochs = 5000
muted = True
batch_size = 64

input_layer_neurons = 2
output_layer_neurons = 2
hidden_layer_neurons = [10, 10]

model_zero_init = MLP(
    input_neurons=input_layer_neurons,
    hidden_layer_neurons=hidden_layer_neurons,
    output_neurons=output_layer_neurons,
    hidden_activation='relu',
    output_activation='softmax',
    loss_function='cross_entropy',
    learning_rate=learning_rate,
    batch_size=batch_size,
    epochs=epochs,
    patience=50
)
model_zero_init.train(X_train, y_train, muted=False, X_val=X_val, y_val=y_val)
# Plot the loss curve
model_zero_init.plot_loss()
```

```python
predictions = model_zero_init.predict(X_val)
true_labels = np.argmax(y_val, axis=1)
accuracy = np.mean(predictions == true_labels)
print(f"Validation Accuracy: {accuracy:.2f}")
```

```python
# Evaluate accuracy on both training and validation sets
model_zero_init.evaluate_accuracy(X_train, y_train, "Training Set")
model_zero_init.evaluate_accuracy(X_val, y_val, "Validation Set")
```
